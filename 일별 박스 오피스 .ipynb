```python
import urllib.request as ul
import json
import csv
import pandas as pd

movieDate = "20200901"
int(movieDate)

url = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=fbba50872f7d4274929b8873a70ac26f&targetDt=20200901"
request = ul.Request(url)

response = ul.urlopen(request)
rescode = response.getcode()

if(rescode == 200):
    responseData = response.read()
    

result = json.loads(responseData)
# result
```

```python
result
```

```python
for i in result['boxOfficeResult']['dailyBoxOfficeList']:
    print(i)
    
data = result['boxOfficeResult']['dailyBoxOfficeList']
data
```

```python
result = pd.DataFrame(data)

result
result = result.rename(columns = {'rnum':"순번", 'rank': '순위', 'rankInten':'순위 증감분', 'rankOldAndNew': '신규진입여부', 'movieCd': '대표코드',
                        'movieNm': '영화명','openDt': '개봉일','salesAmt': '매출액', 'salesShare' : '매출비율' , 'salesInten' : '매출액 증감분',
                        'salesChange':'매출비율', 'salesAcc' : '누적매출액', 'audiCnt' : '관객수',
                        'audiInten' : '관객수 증감분', 'audiChange' : '관객수 증감 비율', 'audiAcc' : '누적관객수', 'scrnCnt' : '스크린수', 'showCnt': '상영된 횟수'})
```
