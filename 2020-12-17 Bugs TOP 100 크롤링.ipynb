```python
import warnings
warnings.filterwarnings('ignore')

from bs4 import BeautifulSoup
from datetime import datetime as dt

import requests
```

```python
# Bugs TOP 100 크롤링하기

url = 'https://music.bugs.co.kr/chart'

# 크롤링 준비
request = requests.get(url)
html = request.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 곡 이름 가져오기

titles = soup.findAll('p', {"class":'title'})

for title in titles:
    print(title.text.strip().split('\n')[0])
    
```

```python
# 아티스트 가져오기

artists = soup.findAll('p', {"class":'artist'})

for artist in artists:
    print(artist.text.strip().split('\n')[0])
    
```

```python
print(dt.now(),'\n')

for i in range(len(titles)):
    title = titles[i].text.strip().split('\n')[0]
    artist = artists[i].text.strip().split('\n')[0]    
    print('{0:3d}위 {1} - {2}'.format(i+1, title, artist))
```

```python
# 크롤링 결과 텍스트 파일에 저장

file = open('BugsTOP100.txt', 'w')

file.write('{} 현재 Bugs 뮤직 실시간 TOP100\n\n'.format(dt.now()))

for i in range(100):
    title = titles[i].text.strip().split('\n')[0]
    artist = artists[i].text.strip().split('\n')[0]
    data = '{0:3d}위 {1} - {2}'.format(i+1, title, artist)
    file.write(data + '\n')

file.close()
```
